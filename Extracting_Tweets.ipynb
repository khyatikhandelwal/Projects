```python
api = tweepy.API(auth,wait_on_rate_limit=True)

tweets_list = tweepy.Cursor(api.search, q="cryptocurrency OR digital currency OR bitcoin OR etherium OR dodgecoin ",
                            geocode = "20.5937,78.9629,1900km", tweet_mode="extended"
                            ).items(10000)
```

```python
output1 = []
for tweet in tweets_list:
    text = tweet._json["full_text"]
    print(text)
    created_at = tweet.created_at
    location = tweet.user._json['geo_enabled']
    
    line = {'text' : text, 'location': location, 'created_at' : created_at}
    output1.append(line)
```

```python
output1 = pd.DataFrame(output1)

print(output1)
```
