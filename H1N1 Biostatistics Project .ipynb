```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df_pandemic = pd.read_csv(r'https://query.data.world/s/gugnzbvij5gwjx3ddklfbhfrvq5dqc',encoding='latin1')
```

```python
df_countrywise_pandemic33 = df_pandemic.pivot_table(index= "Country", values= ["Cases","Deaths"], aggfunc = sum)
df_countrywise_pandemic33.loc["United Kingdom"]
```

```python

df_countrywise_pandemic = df_countrywise_pandemic33[df_countrywise_pandemic33["Cases"] <= 700000]
df_countrywise_pandemic = df_countrywise_pandemic[df_countrywise_pandemic["Cases"] >= 10000]
df_countrywise_pandemic= df_countrywise_pandemic.reset_index()
df_countrywise_pandemic = df_countrywise_pandemic.append({"Country": "United States of America",
                               "Cases": 358659,
                               "Deaths": 1194}, ignore_index= True)
df_countrywise_pandemic = df_countrywise_pandemic.drop([8,9])
df_countrywise_pandemic = df_countrywise_pandemic.reset_index()
df_countrywise_pandemic
```

```python
df_countrywise_pandemic.plot(x="Country", y=["Cases"], kind='bar', color="lightblue", title= "No. of Cases by Country")
```

```python
df_countrywise_pandemic1 = df_pandemic.pivot_table(index= "Country", values= ["Cases","Deaths"], aggfunc = sum)
df_countrywise_pandemic1 = df_countrywise_pandemic1[df_countrywise_pandemic1["Deaths"] != 0]
df_countrywise_pandemic1 = df_countrywise_pandemic1.drop(index= "Grand Total")
df_countrywise_pandemic1 = df_countrywise_pandemic1.reset_index()
df_countrywise_pandemic12 = df_countrywise_pandemic1[df_countrywise_pandemic1["Cases"] >= 34086]
```

```python
fig, ax = plt.subplots(figsize=(16, 7))
sns.set_style("darkgrid")
sns.scatterplot(data= df_countrywise_pandemic12, x="Country", y="Cases", size= "Deaths", color= "black", ax=ax, alpha=0.9, sizes=(400, 1600))
```

```python
df_countrywise_pandemic['deathrate'] = df_countrywise_pandemic["Deaths"]/df_countrywise_pandemic["Cases"]
df_countrywise_pandemic
```

```python
fig, ax1 = plt.subplots(figsize=(9,7))
sns.set_style("white")
df_countrywise_pandemic.plot(x="Country", y=["deathrate"], kind='bar', color="#B0C9C9", ax=ax1, title= "Deathrates by Countries")
```

```python
df_hosp_beds = pd.read_excel(r"https://query.data.world/s/snd5unsj6ym3yoynutmayxnvk3c3rj")
df_hosp_beds
```

```python
df_hosp_beds1 = df_hosp_beds.iloc[:,45:52]
df_hosp_beds1["Country"] = df_hosp_beds["Country Name"]
df_hosp_beds1
```

```python
df_hosp_beds2 = df_hosp_beds1[df_hosp_beds1["Country"] =="Australia"]
df_hosp_beds2
```

```python
df_hosp_beds3 = df_hosp_beds1[df_hosp_beds1["Country"] =="Canada"]
df_hosp_beds3
```

```python
df_hosp_beds4 = df_hosp_beds1[df_hosp_beds1["Country"] =="Chile"]
df_hosp_beds4
```

```python
df_hosp_beds5 = df_hosp_beds1[df_hosp_beds1["Country"] =="Argentina"]
df_hosp_beds5
```

```python
df_hosp_beds6 = df_hosp_beds1[df_hosp_beds1["Country"] =="United Kingdom"]
df_hosp_beds6
```

```python
df_hosp_beds7 = df_hosp_beds1[df_hosp_beds1["Country"] =="Thailand"]
df_hosp_beds7
```

```python
df_hosp_beds8 = df_hosp_beds1[df_hosp_beds1["Country"] =="Mexico"]
df_hosp_beds8
```

```python
df_hosp_beds9 = df_hosp_beds1[df_hosp_beds1["Country"] =="China"]
df_hosp_beds9
```

```python
df_hosp_beds10 = df_hosp_beds1[df_hosp_beds1["Country"] =="Germany"]
df_hosp_beds10 
```

```python
df_final_hosp_beds = df_hosp_beds2.append(df_hosp_beds3)
df_final_hosp_beds = df_final_hosp_beds.append(df_hosp_beds4)
df_final_hosp_beds = df_final_hosp_beds.append(df_hosp_beds5)
df_final_hosp_beds = df_final_hosp_beds.append(df_hosp_beds6)
df_final_hosp_beds = df_final_hosp_beds.append(df_hosp_beds7)
df_final_hosp_beds = df_final_hosp_beds.append(df_hosp_beds8)
df_final_hosp_beds = df_final_hosp_beds.append(df_hosp_beds9)
df_final_hosp_beds = df_final_hosp_beds.append(df_hosp_beds10)

df_final_hosp_beds = df_final_hosp_beds.reset_index()
df_final_hosp_beds = df_final_hosp_beds.drop("index", axis=1)
df_final_hosp_beds_transpose = pd.DataFrame(df_final_hosp_beds.transpose())
df_final_hosp_beds_transpose                    
```

```python
df_final_hosp_beds_transpose[0].fillna(value=1.53, inplace=True)

df_final_hosp_beds_transpose[1].fillna(value=3.22, inplace=True)

df_final_hosp_beds_transpose[2].fillna(value=2.22, inplace=True)

df_final_hosp_beds_transpose[3].fillna(value=4.25, inplace=True)

df_final_hosp_beds_transpose[4].fillna(value=0, inplace=True)

df_final_hosp_beds_transpose[5].fillna(value=2.1, inplace=True)

df_final_hosp_beds_transpose[6].fillna(value=1.53, inplace=True)

df_final_hosp_beds_transpose[7].fillna(value=3.01, inplace=True)

df_final_hosp_beds_transpose[8].fillna(value=0, inplace=True)
```

```python
df_final_hosp_beds = pd.DataFrame(df_final_hosp_beds_transpose.transpose())
df_final_hosp_beds = df_final_hosp_beds.drop(["2004","2005","2006"], axis=1)
df_final_hosp_beds
```

```python
filename = "HospBeds.xlsx"
df_final_hosp_beds.to_excel(filename)
```
